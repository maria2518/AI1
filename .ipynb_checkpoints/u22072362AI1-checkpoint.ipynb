### Servidor

```python
import socket
import threading
import sys
import pickle
import os
#Para que el usuario pueda introducir el puerto usamos la función input y a continuación la insrucción que 
#deberá de seguir el usuario para introducir los datos que seran guardados en la variable "puero". 
#En este caso con el int determinamos que los datos introducidos tienen que ser de tipo integer, es decir numeros enteros.
class Servidor():
	def __init__(self, host=socket.gethostname(), puerto=int(input("Introduzca el puerto que desea utilizar: "))):
		self.clientes = []
#En la siguiente línea de codigo lo que estoy haciendo es imprimir la dirección IP mediante 
#la función socket.gethostbyname a la que le paso la variable "host" que es donde se ha guardado la dirección
#a la que se ha conectado el servido anteriormente. 
		print("\n Su IP es: ",socket.gethostbyname(host))
#Con este print lo hago es imprimir el puerto que el usuario introdujo con aterioridad
		print("\n Su puerto es: ", puerto, "\n")
		self.sock = socket.socket()
		self.sock.bind((str(host), int(puerto)))
		self.sock.listen(20)
		self.sock.setblocking(False)

		aceptar = threading.Thread(target=self.aceptarC)
		procesar = threading.Thread(target=self.procesarC)

		aceptar.daemon = True
		aceptar.start()

		procesar.daemon = True
		procesar.start()

		while True:
			msg = input('SALIR = Q\n')
			if msg == 'Q':
				print("Hasta luego, vuelva pronto")
				self.sock.close()
				sys.exit()
			else:
				pass

	def broadcast(self, msg, cliente):
		for c in self.clientes:
			try:
				if c != cliente:
					c.send(msg)
			except:
				self.clientes.remove(c)

	def aceptarC(self):
		while True:
			try:
				conn, addr = self.sock.accept()
				print(f"\nConexion aceptada via {conn}\n")
				conn.setblocking(False)
				self.clientes.append(conn)
			except:
				pass
#Para que se puedan guardar un registro de los mensajes enviados y el usuario que los envia, en un txt, lo primero
#que hago es crear una función escribir, que como su nombre indica se encargara de escribir los datos en el fichero.
#Para ello le pasamos la variable msg que se corresponderá con los datos que queremos guardar. En la siguiente línea
#lo que hago es guardar en la variable file mediante la función open un fichero con el nombre indicado, 
#que en el caso de que esté creado solo abrirá el fichero, pero en caso contrario primero lo creara.
#A continuación con file.write lo que hago es escribir los datos que le he pasado en la variable msg, y por último
#cerramos el fichero.             
	def escribir(self, msg):
		file=open("u22072362AI1.txt", "a")
		file.write(msg + "\n")
		file.close()

	def procesarC(self):
		print("Procesamiento de mensajes iniciado")
		while True:
			if len(self.clientes) > 0:
				for c in self.clientes:
					try:
						data = c.recv(32)
						if data:
							self.broadcast(data,c)
#LLamo a la función escribir y le paso todos los datos don la funcion pickle.loads que me permite traducir 
#los bloques de bytes y pasarlos a objetos.
							self.escribir(pickle.loads(data))
					except:
						pass

s = Servidor()
```

### Cliente

```python
import threading
import sys
import socket
import pickle
import os

class Cliente():
    
#Para que el cliente pueda establecer conxión con el servidor deverá introducir los MISMOS datos (IP y puero)
#dados por el servidor y estos son recogidos a traves del input en sus correspondientes variables'''

	def __init__(host=input("Introduzca la IP dada por el servidor: "), puerto=int(input("Introduzca el puerto del servidor: ")), nombre=input("Introduzca su nombre: ")):
		self.nombre = nombre
		self.sock = socket.socket()
		self.sock.connect((str(host), int(puerto)))
		hilo_recv_mensaje = threading.Thread(target=self.recibir)
		hilo_recv_mensaje.daemon = True
		hilo_recv_mensaje.start()
		print('Hilo con PID',os.getpid())
		print('Hilos activos', threading.active_count())

		while True:
			msg = input('\nEscriba texto ? ** Enviar = ENTER ** Abandonar Chat = Q \n')
			if msg != 'Q' :
#Para que sea visible el nombre junto con el mensage que envia cada cliente lo que hago es pasarle a la función 
#self.enviar la variable nombre en la que se guardó el nickname que el usuario introdujo y el mensage enviadopor el usuario'''
				self.enviar(nombre + ": " + msg)
			else:
				print("Hasta luego")
				self.sock.close()
				sys.exit()

	def recibir(self):
		while True:
			try:
				data = self.sock.recv(32)
				if data:
					print(pickle.loads(data))
			except:
				pass

	def enviar(self, msg):
		self.sock.send(pickle.dumps(msg))

c = Cliente()
```
